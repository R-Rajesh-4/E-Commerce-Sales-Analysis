```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset= pd.read_csv("./SuperStoreUS.csv")
```

```python
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'], format='%d/%m/%Y')

 
dataset["Order Year"] =dataset["Order Date"].dt.year 
```

```python
dataset.shape
```

```python
#dataset.isnull().sum()
```

```python
#after running this we get 0 in above cell
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(), inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Product Category', y='Product Base Margin', data=dataset, estimator=sum)
plt.show()
```

```python
dataset['State or Province'].value_counts()[:5]
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Product Category', y='Profit', data=dataset, estimator=sum)
plt.show()
```

# Product Category

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Product Category', data=dataset)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Customer Segment', data=dataset[dataset["Product Category"]=="Office Supplies"], hue="Product Sub-Category")
plt.show()
```
